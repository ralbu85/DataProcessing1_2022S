## 사각형의 면적 구하는 프로그램

```python
# 데이터
h = 10
v = 20

# 데이터를 조작 --> 면적이라는 새로운 데이터 생성
def area(x,y):
    return x*y

# 결과 확인
area(h,v)
```

## 클래스 정의

```python
class Rectangle():
    def __init__(self, x,y):
        self.h = x
        self.v = y
    
    def area(self):
        return self.h * self.v      
```

## 클래스로부터 객체만들기

```python
r = Rectangle(10,20) #데이터 받기
a = r.area() #받은 데이터를 이용해 새로운 값 만들기
print(a)
```

```python
a = Rectangle(1, 1)   # 가로 1, 세로 1인 사각형
b = Rectangle(2, 1)   # 가로 2, 세로 1인 사각형
c = Rectangle(4, 2)   # 가로 4, 세로 2인 사각형
d = Rectangle(6, 3)   # 가로 6, 세로 3인 사각형
e = Rectangle(8, 5)   # 가로 8, 세로 5인 사각형
```

```python
print(a.area())
print(b.area())
print(c.area())
print(d.area())
print(e.area())
```

## 캐릭터 클래스

```python
class Character():
    def __init__(self):
        self.life = 1000
        
    def attacked(self):
        self.life = self.life - 10
        print("공격받음! 생명력 = ", self.life)
```

```python
a = Character()
b = Character()
c = Character()
```

```python
a.attacked()
```

```python
for i in range(3):
    b.attacked()
```

```python
for i in range(10):
    a.attacked()
```

## 클래스 상속

```python
class Warrior(Character):
    def __init__(self):
        super(Warrior, self).__init__()
        self.strength = 15
        self.intelligence = 5        
```

```python
class Wizard(Character):
    def __init__(self):
        super(Wizard, self).__init__()
        self.strength = 5
        self.intelligence = 15      
```

```python
a = Warrior()
b = Wizard()
```

```python
a.life, b.life
```

```python
a.strength, b.strength
```

```python
a.intelligence, b.intelligence
```

```python
a.attacked()
```

```python
b.attacked()
```

## 메소드 오버라이딩

```python
class Warrior(Character):
    def __init__(self):
        super(Warrior, self).__init__()
        self.strength = 15
        self.intelligence = 5
        
    def attack(self):
        print("육탄공격")
```

```python
class Wizard(Character):
    def __init__(self):
        super(Wizard, self).__init__()
        self.strength = 5
        self.intelligence = 15
        
    def attack(self):
        print('원거리 공격')
```

```python
a = Warrior()
b = Wizard()
a.attack()
b.attack()
```
