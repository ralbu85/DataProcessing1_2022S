## 불리언 배열

```python
idx = np.array([True, False, False, True, True])
idx
```

```python
a=np.array([0,1,2,3,4])
```

```python
a[idx]
```

## 팬시 인덱싱(조건부 인덱싱)

```python
a>2
```

```python
a[a>2]
```

```python
a%2==0
```

```python
a[a%2==0]
```
